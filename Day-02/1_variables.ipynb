## 1. Variables are containers that store data in a Python program

### 2. Variables can store different types of data types such as string, integer, float, boolean.

### 3. Rules of naming variables correctly:

- you cannot use reserve words as variable names(e.g. def, True etc.).
- Variables names must start with a letter or an underscore _  .
- It cannot contain spaces or special characters(@,#,$,%, etc.).
- Underscore is a valid character and can be present at any place in a variable name
