                                    ##  Takeaways  ##
- Integer numbers store whole numbers (without decimal part), e.g. 57.
- Float numbers store fractional numbers with whole and decimal part, e.g. 57.23.
- type(variable_name) can be used to detect the data type of a variable.
- / operator is used for division whereas // is used to retrieve integer part of the division.
- % is a modulo operator, it returns the remainder of a division operator.
