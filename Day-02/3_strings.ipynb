                                        ## Takeaways ##

- In python strings are immutable, they can't be changed once created.

- Access specific characters or substrings in python using indexing(e.g.,name[0]) and slicing(e.g.,name[1:5]).

- Use string formatting techniques like f-strings (e.g.,f'{variable}') for easier and more readable string composition in python.

- Python provides built-in methods for strings, such as .upper(), .lower(), .split(), and .strip().
